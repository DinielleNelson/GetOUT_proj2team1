```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://en.wikipedia.org/wiki/Comparison_of_crewed_space_vehicles'
response = requests.get(url)
```

```python
soup = bs(response.text,'lxml')
```

```python
print(soup.prettify())
```

```python
# retrieves all tables in page
tables = soup.body.find_all('table').text
tables
```

```python
# Retreive second table - Sub-Orbital Spacecraft
table_a = tables[0]
table_a
```
